```python
print("LOADING OLIST RAW FILES")
print("=" * 55)

def load_file(filename):
    path = DATA_PATH + filename
    try:
        df = pd.read_csv(path)
        print(f"{filename:<45} {len(df):>8,} rows")
        return df
    except FileNotFoundError:
        print(f" NOT FOUND: {filename}")
        return None

orders      = load_file("olist_orders_dataset.csv")
items       = load_file("olist_order_items_dataset.csv")
payments    = load_file("olist_order_payments_dataset.csv")
reviews     = load_file("olist_order_reviews_dataset.csv")
customers   = load_file("olist_customers_dataset.csv")
sellers     = load_file("olist_sellers_dataset.csv")
products    = load_file("olist_products_dataset.csv")
translation = load_file("product_category_name_translation.csv")
geo         = load_file("olist_geolocation_dataset.csv")

print("\n All files loaded!")
```

```python
# Parse all date columns
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

print("Dates parsed")
print(orders[date_cols].dtypes)
```

```python
# Filter to delivered orders only
orders_delivered = orders[orders['order_status'] == 'delivered'].copy()
print(f"Total orders:     {len(orders):,}")
print(f"Delivered orders: {len(orders_delivered):,}")
print(f"Other statuses:   {orders['order_status'].value_counts().to_dict()}")
```

```python
# ── Delivery timing metrics ───────────────────────────────────
orders_delivered['delay_days'] = (
    orders_delivered['order_delivered_customer_date'] -
    orders_delivered['order_estimated_delivery_date']
).dt.days

orders_delivered['actual_delivery_days'] = (
    orders_delivered['order_delivered_customer_date'] -
    orders_delivered['order_purchase_timestamp']
).dt.days

# Binary flags — mutually exclusive, must sum to 100%
orders_delivered['is_late']      = (orders_delivered['delay_days'] > 0).astype(int)
orders_delivered['on_time_flag'] = (orders_delivered['delay_days'] <= 0).astype(int)

# ── Delay severity buckets ────────────────────────────────────
# Using pd.cut for performance on large datasets
# Negative delay = delivered early (positive for customer experience)
orders_delivered['delay_severity'] = pd.cut(
    orders_delivered['delay_days'],
    bins      = [-float('inf'), 0, 3, 7, float('inf')],
    labels    = ['On Time', 'Slightly Late (1-3d)', 
                 'Moderately Late (4-7d)', 'Severely Late (7d+)'],
    right     = True
).astype(str)

# ── Date dimensions for Power BI slicing ─────────────────────
orders_delivered['order_year']      = orders_delivered['order_purchase_timestamp'].dt.year
orders_delivered['order_month']     = orders_delivered['order_purchase_timestamp'].dt.month
orders_delivered['order_quarter']   = orders_delivered['order_purchase_timestamp'].dt.quarter
orders_delivered['order_month_str'] = orders_delivered['order_purchase_timestamp'].dt.to_period('M').astype(str)

# ── Validation: on_time + late must equal total ───────────────
total       = len(orders_delivered)
on_time     = orders_delivered['on_time_flag'].sum()
late        = orders_delivered['is_late'].sum()
early       = (orders_delivered['delay_days'] < 0).sum()

assert on_time + late == total, "FLAG ERROR: on_time + late != total orders"

otd_rate    = on_time / total * 100
late_rate   = late / total * 100
avg_delay   = orders_delivered.loc[orders_delivered['is_late']==1, 'delay_days'].mean()
avg_early   = abs(orders_delivered.loc[orders_delivered['delay_days']<0, 'delay_days'].mean())

print("Delivery Feature Engineering — Complete")
print(f"  Orders analyzed:       {total:>10,}")
print(f"  On-time deliveries:    {on_time:>10,}  ({otd_rate:.1f}%)")
print(f"  Late deliveries:       {late:>10,}  ({late_rate:.1f}%)")
print(f"  Delivered early:       {early:>10,}  (avg {avg_early:.1f} days early)")
print(f"  Avg delay (late only): {avg_delay:>10.1f} days")
print()

# ── Target gap ────────────────────────────────────────────────
TARGET_OTD = 92.0
gap = otd_rate - TARGET_OTD
status = "✓ Above target" if gap >= 0 else f"✗ {abs(gap):.1f} pts below target"
print(f"  OTD vs 92% target:     {status}")

print()
print("Delay severity distribution:")
print(orders_delivered['delay_severity'].value_counts().to_string())
```

```python

base_count = len(orders_delivered)

# ── Items: revenue + freight + seller per order ───────────────
# primary_seller_id = first seller listed on the order
# For multi-seller orders this is a simplification — acceptable
# given 97%+ of Olist orders have a single seller
items_agg = items.groupby('order_id').agg(
    total_revenue     = ('price',            'sum'),
    total_freight     = ('freight_value',    'sum'),
    item_count        = ('order_item_id',    'count'),
    primary_seller_id = ('seller_id',        'first'),
    product_id        = ('product_id',       'first')
).reset_index()

# ── Payments: total value + method per order ──────────────────
payments_agg = payments.groupby('order_id').agg(
    total_payment = ('payment_value',        'sum'),
    payment_type  = ('payment_type',         'first'),
    installments  = ('payment_installments', 'max')
).reset_index()

# ── Reviews: one review per order (keep most recent if multiple)
reviews_agg = (
    reviews
    .sort_values('review_creation_date', ascending=False)
    .drop_duplicates(subset='order_id')
    [['order_id', 'review_score', 'review_creation_date']]
)

# ── Build fact table ──────────────────────────────────────────
fact = (
    orders_delivered
    .merge(items_agg,    on='order_id', how='left')
    .merge(payments_agg, on='order_id', how='left')
    .merge(reviews_agg,  on='order_id', how='left')
    .merge(
        customers[['customer_id', 'customer_state', 'customer_city']],
        on='customer_id', how='left'
    )
)

# ── Validate: merges must not create or lose rows ─────────────
assert len(fact) == base_count, \
    f"ROW COUNT ERROR: started with {base_count:,}, ended with {len(fact):,}"

dupe_count = fact['order_id'].duplicated().sum()
assert dupe_count == 0, \
    f"DUPLICATE ERROR: {dupe_count:,} duplicate order_ids in fact table"

# ── Summary ───────────────────────────────────────────────────
review_coverage = fact['review_score'].notna().mean() * 100
seller_coverage = fact['primary_seller_id'].notna().mean() * 100
revenue_nulls   = fact['total_revenue'].isna().sum()

print(f"✓ fact_orders built: {len(fact):,} rows × {len(fact.columns)} columns")
print()
print(f"  Coverage checks:")
print(f"  Review score populated:   {review_coverage:.1f}%")
print(f"  Seller ID populated:      {seller_coverage:.1f}%")
print(f"  Revenue nulls:            {revenue_nulls:,}")
print()
print(f"  Columns ({len(fact.columns)}):")
for i, col in enumerate(fact.columns, 1):
    print(f"    {i:>2}. {col}")
```

```python
# dim_sellers — add state info
dim_sellers = sellers.rename(columns={
    'seller_id': 'seller_id',
    'seller_zip_code_prefix': 'seller_zip',
    'seller_city': 'seller_city',
    'seller_state': 'seller_state'
}).copy()
print(f" dim_sellers: {len(dim_sellers):,} rows")

# dim_products — translate category names
dim_products = products.merge(
    translation, on='product_category_name', how='left'
).copy()
dim_products['category_english'] = dim_products['product_category_name_english'].fillna(
    dim_products['product_category_name']
)
print(f" dim_products: {len(dim_products):,} rows")

# dim_customers
dim_customers = customers.copy()
print(f" dim_customers: {len(dim_customers):,} rows")

# dim_geolocation — deduplicate to one row per zip
dim_geo = geo.groupby('geolocation_zip_code_prefix').agg(
    lat   = ('geolocation_lat', 'mean'),
    lng   = ('geolocation_lng', 'mean'),
    city  = ('geolocation_city', 'first'),
    state = ('geolocation_state', 'first')
).reset_index()
print(f" dim_geolocation: {len(dim_geo):,} rows")
```

```python
# ── CELL: Fix the 8 nulls before loading ──────────────────

print("Before fix:")
print(f"  delay_days nulls:   {fact['delay_days'].isnull().sum()}")
print(f"  review_score nulls: {fact['review_score'].isnull().sum()}")

# Fix 1: delay_days — 8 rows with missing actual delivery date
# These orders were delivered but delivery date wasn't recorded
# Safe to set delay = 0 (treat as on-time, unknown delay)
fact['delay_days'] = fact['delay_days'].fillna(0)

# Fix 2: review_score — 646 orders with no review submitted
# Use -1 as a sentinel so Power BI can filter them out separately
# (do NOT use 0 — that would pull down averages incorrectly)
fact['review_score_raw'] = fact['review_score'].copy()   # keep original
fact['review_score']     = fact['review_score'].fillna(-1)
fact['has_review']       = (fact['review_score'] > 0).astype(int)

# Recheck
print("\nAfter fix:")
print(f"  delay_days nulls:   {fact['delay_days'].isnull().sum()}")
print(f"  review_score nulls: {fact['review_score'].isnull().sum()}")

# Quick sanity check
print(f"\n  Total rows:                {len(fact):,}")
print(f"  Rows with review:          {fact['has_review'].sum():,}")
print(f"  Rows without review:       {(fact['has_review']==0).sum():,}")
print(f"  delay_days range:          {fact['delay_days'].min()} → {fact['delay_days'].max()}")

print("\n All nulls handled — safe to load to SQL Server")
```
